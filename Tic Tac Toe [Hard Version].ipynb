## Create Our Board 
`board` will be an empty of values `_`

we will choose to computer to play with character `x` and user will play with character `o`

```python
# board with shape(3: row, 3: col)
board = [
    ['_', '_', '_'],
    ['_', '_', '_'],
    ['_', '_', '_']
]

# initialize Our Players 
player, opponent = 'x', 'o'
```

## empty_board() function

```python
def empty_board():
    global board
    board = [["_"] * 3 for _ in range(3)]
```

## show_board(b) function

```python
# print Our board 
def show_board(b):
    # rows
    for i in range(3):
        # cols
        for j in range(3):
            
            if b[i][j] == 'x':
                print("\033[92;1mx \033[90;0m", end = "")
            elif b[i][j] == 'o':
                print('\033[94;1mo \033[90;0m', end = "")
            else:
                print("_ ", end = "")
        # newline
        print()
```

### Check Are There's Another Moves 
to do this we should check if there's en empty cell that doesn't contain any value of `x or o`

## isMovesLeft(board) function

```python
def isMovesLeft(board) :
    # iteration through row
    for i in range(3) :
        # iteraiton through col
        for j in range(3) :
            # check if there's empty cell that doesn't contain x or o
            if (board[i][j] == '_') :
                return True
            
    return False
```

### Evaluation to Check If There's A Winner

check if any of the `columns` or `rows` values are equal 

          In Row board[1][1] == board[1][2] == board[1][3]
          In Col board[1][1] == board[2][1] == board[3][1]
          
          and so on...
          
          check also in diagonals
          
          diag_1 => board[1][1] == board[2][2] == board[3][3]
          diag_2 => board[1][3] == board[2][2] == board[3][1]
          

#### In  This Example
| Board | col_1 | col_2 | col_3 |
|------|-----|------------| ----- |
| row_1 | x | o | o |
| row_2 | o | x | x |
| row_3 | o | x | x |

The winner is `x` because the main diagonal values are The same. 

## evaluate(b) function

```python
def evaluate(b) :
    """
        Params: 
                b: board
        Returns: 
                10 or -10 or 0
                10: player => "x"
                -10: player => "o"
                0: player => no winner
    """
    # for rows values
    # board[row][0] == board[row][1] == board[row][2]
    for row in range(3) :
        if (b[row][0] == b[row][1] and b[row][1] == b[row][2]) :
            if (b[row][0] == player) : # x
                return 10
            elif (b[row][0] == opponent) : # o
                return -10

    # for cols values
    # board[0][col] == board[1][col] == board[2][col]
    for col in range(3) :

        if (b[0][col] == b[1][col] and b[1][col] == b[2][col]) :

            if (b[0][col] == player) :
                return 10
            elif (b[0][col] == opponent) :
                return -10

    # main Diagonal
    if (b[0][0] == b[1][1] and b[1][1] == b[2][2]) :

        if (b[0][0] == player) :
            return 10
        elif (b[0][0] == opponent) :
            return -10

    # sub diagonal
    if (b[0][2] == b[1][1] and b[1][1] == b[2][0]) :

        if (b[0][2] == player) :
            return 10
        elif (b[0][2] == opponent) :
            return -10

    return 0
```

## Minimax Algorithm

`Minimax` is a kind of `backtracking` algorithm that is used in decision making and game theory to find the optimal move for a player, assuming that your opponent also plays optimally. 

It is widely used in two player turn-based games such as `Tic-Tac-Toe`, `Backgammon`, `Mancala`, `Chess`, `etc`.
In `Minimax` the two players are called `maximizer` and `minimizer`. 

`The Maximizer` tries to get the highest score possible while `the minimizer` tries to do the opposite and get the lowest score possible.

Every board state has a value associated with it. 

In a given state if `the maximizer` has upper hand then, the score of the board will tend to be some positive value. 

If `the minimizer` has the upper hand in that board state then it will tend to be some negative value. 

The values of the board are calculated by some heuristics which are unique for every type of game.


Example: 
Consider a game which has **4** final states and paths to reach final state are from root to 4 leaves of a perfect binary tree as shown below. 

Assume you are the maximizing player and you get the first chance to move, i.e., you are at the root and your opponent at next level. 

**Which move you would make as a maximizing player considering that your opponent also plays optimally?**

Since this is a `backtracking` based algorithm, it tries all possible moves, then backtracks and makes a decision. 

`Maximizer` goes **LEFT**: It is now `the minimizers` turn. 
`The minimizer` now has a choice between **3** and **5**. 
Being the `minimizer` it will definitely choose the least among both, that is **3**


`Maximizer` goes **RIGHT**: It is now `the minimizers` turn. `The minimizer` now has a choice between **2** and **9**. He will choose **2** as it is the least among the two values.

Being `the maximizer` you would choose the larger value that is **3**. 

Hence the optimal move for `the maximizer` is to go **LEFT** and the optimal value is **3**.

## minimax(board, depth, isMax) function

```python
def minimax(board, depth, isMax) :
    # score = [0, 10, -10]
    score = evaluate(board)

    # base case
    # return score if score == [10, -10]
    if (score == 10) :
        return score


    # base case
    if (score == -10) :
        return score

    # base case
    # return 0 if there's aren't any empty cell in our board
    if (isMovesLeft(board) == False) :
        return 0

    # Take Turns between Maximizer and Minimizer
    
    if (isMax) :
        
        # negative infinity value 
        best = -math.inf

        # check for every value empty cell in board
        for i in range(3) :
            for j in range(3) :

                if (board[i][j]=='_') :
                    # board[i][j] = 'x'
                    board[i][j] = player

                    # backtrack and get the max value of (-math.inf, minimax(....))
                    best = max(best, minimax(board,
                                            depth + 1,
                                            not isMax) )

                    board[i][j] = '_'
        return best

    # reverse but for o
    else :
        best = math.inf

        for i in range(3) :
            for j in range(3) :

                if (board[i][j] == '_') :

                    board[i][j] = opponent

                    best = min(best, minimax(board, depth + 1, not isMax))

                    board[i][j] = '_'
        return best
```

### findBestMove(board) function

```python
def findBestMove(board) :
    bestVal = -math.inf
    bestMove = (-1, -1)
    
    # every empty cell in board
    for i in range(3) :
        for j in range(3) :

            if (board[i][j] == '_') :

                # assume an empty cell with x
                # check if this will give me the highest value
                board[i][j] = player

                moveVal = minimax(board, 0, False)

                # undo
                board[i][j] = '_'

                # Get The highest value
                if (moveVal > bestVal) :
                    # store in bestMove that will be return in shape of (i, j)
                    bestMove = (i, j)
                    # change bestVal
                    bestVal = moveVal
    # bestMove
    return bestMove
```

### start_game() Function

```python
def start_game():
    
    # print cells in numbers
    print("\033[90;0m")
    for i in range(9):
        print(i+1, end = " ")
        if (i + 1) % 3 == 0:
            print()
    print("------------------------")
    while isMovesLeft(board):
        user_cell = input("Enter Cell (1-9): ")
        if user_cell.isdigit():

            user_cell = int(user_cell) - 1

            if user_cell < 0 or user_cell > 8 or board[user_cell // 3][user_cell % 3] != '_':
                print("\033[91mInvalid Cell Entering\033[90m")
                continue
            else:
                board[user_cell // 3][user_cell % 3] = 'o'
                print("\033[1m----------\033[0m")

                show_board(board)
                
                winner = evaluate(board)
                if winner == -10:
                    print("\033[94;1mO Wins!")
                    break
        else:
            print("Enter Number Not A String.")
            continue

        if isMovesLeft(board) == True:
            best_move_for_computer = findBestMove(board)
            board[best_move_for_computer[0]][best_move_for_computer[1]] = 'x'

            winner = evaluate(board)
            if winner == 10:
                show_board(board)
                print("\033[1m----------\033[0m")
                print("\033[92;1mX Wins!")
                break
            print("\033[1m----------\033[0m")
            show_board(board)
    else:
        print("\033[97;1mTie!")
    
```
